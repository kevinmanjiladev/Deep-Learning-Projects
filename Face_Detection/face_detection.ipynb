```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('image.png')
```

```python
color_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(color_image)
```

```python
# face detection
# CascadeClassifier is a class used to pass the model file

face_data=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_data
```

```python
# detectMultiScale is a function used to detect the face
faces=face_data.detectMultiScale(color_image,minNeighbors=5)
faces
# minNeighbors is a parameter used in Haar Cascade detection (like face detection) in OpenCV. 
# It helps control how strict the detector is when deciding whether something is truly a face/object.
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(color_image,(x,y),(x+w,y+h),(255,0,0),3)
    
```
