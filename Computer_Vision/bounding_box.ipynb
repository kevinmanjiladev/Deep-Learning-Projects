```python
import cv2
image=cv2.imread('face1.jpg')
image
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
color_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(color_image)
```

How to plot RECTANGLE on a face
---

```python
# To draw rectangle on a face: 
# we need 2 points=>Left Top and Right Bottom
# Total cordinates=>4


cv2.rectangle(color_image,pt1=(350,120),pt2=(600,490),color=(255,0,0),thickness=3)
plt.imshow(color_image)
```

```python
image2=cv2.imread('face2.jpg')
color_image2=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
# plt.imshow(color_image2)
cv2.rectangle(color_image2,pt1=(200,75),pt2=(360,290),color=(255,0,0),thickness=3)
plt.imshow(color_image2)
```

How to plot CIRCLE on a face
---

```python
image3=cv2.imread('face1.jpg')
color_image3=cv2.cvtColor(image3,cv2.COLOR_BGR2RGB)
plt.imshow(color_image3)
cv2.circle(color_image3,center=(480,290),radius=180,color=(0,0,255),thickness=3)
plt.imshow(color_image3)
```

How to add TEXT in an image:
---

    we use puttext paramter 

```python
image4=cv2.imread('face1.jpg')
color_image4=cv2.cvtColor(image4,cv2.COLOR_BGR2RGB)
cv2.putText(color_image4,text='Versatile Actor',org=(700,200),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=1,color=(0,0,150),thickness=2)
plt.imshow(color_image4)
```
