OPEN CV for performing the computer vision operations

```python
import cv2
```

```python
# How to read an image

image=cv2.imread('face1.jpg')
```

```python
image 
# Image read in pixel values
```

```python
# how to display an image

cv2.imshow("Frame",image)  # to display the image
cv2.waitKey(10000)  # image will opened in a seperate window called FRAME for 10 seconds
cv2.destroyAllWindows()  # window will be closed after 10 seconds
```

```python
image2=cv2.imread('face2.jpg')
```

```python
cv2.imshow("Frame2",image2)
cv2.waitKey(10000) # 1000= 1second, 10000=10 seconds
cv2.destroyAllWindows()
```

```python
# how to convert image into gray scale

gray_img=cv2.imread("face3.jpg",cv2.IMREAD_GRAYSCALE)
cv2.imshow("Frame3",gray_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
# How to save the gray scale image

cv2.imwrite('face3_grayscale.jpg',gray_img)
```

```python
gray_img
```

```python
# # How to convert image to binary

# # To convert image into binary: first we should convert color image to gray scale and then to binary image

# # here we use THRESH_BINARY to convert gray scale image into binary

# the result will produce 2 variables

thresh,bin_image=cv2.threshold(gray_img,140,255,cv2.THRESH_BINARY)
bin_image
```

```python
cv2.imshow("Frame4",bin_image)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
thresh
```

```python
# we can also use THRESH_OTSU for conversion of gray scale image to binary

thres1,bin_img1=cv2.threshold(gray_img,100,255,cv2.THRESH_OTSU)
cv2.imshow("Frame5",bin_img1)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
thres1
```

```python
# THRESH_OTSU automatically assign best value for thresh1
```

We display the image using opencv(imshow),but we need x and y cordinates of images.so we should use matplotlib for identify x and y cordinates

```python
img1=cv2.imread('face1.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
```

The above image is in BGR format. so first we should convert BGR to RGB. And then use matplotlib to diplay the image

```python
converted_img=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```
