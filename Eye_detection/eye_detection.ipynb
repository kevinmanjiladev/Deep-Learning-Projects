```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('image.png')
```

```python
plt.imshow(image)
```

```python
color_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
color_image
```

```python
plt.imshow(color_image)
```

```python
model=cv2.CascadeClassifier('haarcascade_eye.xml')
model
```

```python
eyes=model.detectMultiScale(color_image,minNeighbors=4)
eyes
```

```python
for(x,y,w,h) in eyes:
    cv2.rectangle(color_image,(x,y),(x+w,y+h),(0,0,255),3)
```

```python
plt.imshow(color_image)
```
