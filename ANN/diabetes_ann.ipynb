```python
from sklearn.metrics import classification_report

def ANN(x_train,y_train,x_test,y_test,loss,weights):
    model=keras.Sequential([
        keras.layers.Dense(8,activation='relu'),
        keras.layers.Dense(10,activation='relu'),
        keras.layers.Dense(6,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ])

    model.compile(optimizer='adam',loss=loss,metrics=['accuracy'])

    if weights==-1:
        model.fit(x_train,y_train,epochs=200)
    else:
        model.fit(x_train,y_train,epochs=100,class_weight=weights)
    
    print(model.evaluate(x_test,y_test))

    y_pred=model.predict(x_test)
    y_pred=np.round(y_pred)

    print("Classifaction report: \n",classification_report(y_test,y_pred))
    return y_pred

```

```python
y_pred1=ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
y_pred1.reshape(-1)
```

```python
x_ensemble=df.iloc[:,:-1]
y_ensemble=df.iloc[:,-1]
```

```python
y_ensemble.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train_ensm,x_test_ensm,y_train_ensm,y_test_ensm=train_test_split(x_ensemble,y_ensemble,test_size=0.2,random_state=42,stratify=y_ensemble)
```

```python
df_class_0_ensemble=df.loc[df['Outcome']==0]
df_class_1_ensemble=df.loc[df['Outcome']==1]
df_class_0_ensemble['Outcome'].value_counts()
df_class_1_ensemble['Outcome'].value_counts()
# 500/268
# so,df_class_0_ensemble is splitted into 1.8656716417910448 batches
```

```python
df_class_0_ensemble.shape,df_class_1_ensemble.shape
```

```python
df_class_0_ensemble[:268].shape #here we are taking only 268 elements from class_0(slicing techinque is used here)
# so now ,class_0 and class_1 have equal number of 0 and 1
```

```python
df_ensemble=pd.concat([df_class_0_ensemble[:268],df_class_1_ensemble],axis=0)
df_ensemble
```

```python
df_ensemble['Outcome'].value_counts()
```

```python
def get_batch(df_majority,df_minority,start,end):
    df_train=pd.concat([df_majority[start:end],df_minority],axis=0)
    x=df_train.iloc[:,:-1]
    y=df_train.iloc[:,-1]

    from sklearn.model_selection import train_test_split
    x_train_ensm,x_test_ensm,y_train_ensm,y_test_ensm=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
    
    return x_train_ensm,y_train_ensm,x_test_ensm,y_test_ensm
```

```python
x_train_ensm1,y_train_ensm1,x_test_ensm1,y_test_ensm1=get_batch(df_class_0_ensemble,df_class_1_ensemble,0,268)
y_train_ensm1.value_counts()
```

```python
y_test_ensm1.value_counts()
```

```python
y_pred_ensm1=ANN(x_train_ensm1,y_train_ensm1,x_test_ensm1,y_test_ensm1,'binary_crossentropy',-1)
y_pred_ensm1.reshape(-1)
```

```python
x_train_ensm1.shape[0]+x_test_ensm1.shape[0]
```

```python
x_train_ensm2,y_train_ensm2,x_test_ensm2,y_test_ensm2=get_batch(df_class_0_ensemble,df_class_1_ensemble,268,536)
y_pred_ensm2=ANN(x_train_ensm2,y_train_ensm2,x_test_ensm2,y_test_ensm2,'binary_crossentropy',-1)
y_pred_ensm2.reshape(-1)
```
