A neuron in a neural network does three things:

    Takes inputs ==>[x1,x2,x3,x4]

    Applies weights + bias ==>   y=[M1*X1+M2*X2+M3*X3+M4*X4]+C

    Passes the result through an activation function  ==> (1/1+e^-y)

    Outputs a prediction  ==> values between 0 and 1

Logistic regression does exactly the same.
