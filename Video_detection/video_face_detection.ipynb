Face detection in a video
---

```python
import cv2
model=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
video=cv2.VideoCapture('internal_video.mp4')
while True:
    success,img = video.read()
    if not success or img is None:
        print("Video ended.")
        break
    img_small = cv2.resize(img, (640, 360))
    gray_img=cv2.cvtColor(img_small,cv2.COLOR_BGR2GRAY)
    faces=model.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
    print(faces)
    for (x,y,w,h) in faces:
        cv2.rectangle(img_small,(x,y),(x+w,y+h),(0,0,255),3)

    cv2.imshow('Frame',img_small)
    if cv2.waitKey(10)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
