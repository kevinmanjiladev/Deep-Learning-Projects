```python
df['Age'] = df['Age'].fillna(df['Age'].mean()).round(2)
```

```python
df['Ever_Married']=df['Ever_Married'].fillna(df['Ever_Married'].mode()[0])
```

```python
df['Graduated']=df['Graduated'].fillna(df['Graduated'].mode()[0])
```

```python
df['Profession']=df['Profession'].fillna(df['Profession'].mode()[0])
```

```python
df['Work_Experience']=df['Work_Experience'].fillna(round(df['Work_Experience'].mean()))
```

```python
df['Family_Size']=df['Family_Size'].fillna(round(df['Family_Size'].mean()))
```

```python
df['Var_1']=df['Var_1'].fillna(df['Var_1'].mode()[0])
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
cols=['Gender','Ever_Married','Graduated','Segmentation']
for i in cols:
    df[i]=lab.fit_transform(df[i])
df
```

```python
df=pd.get_dummies(df,columns=['Profession','Spending_Score','Var_1'],drop_first=True,dtype=int)
```

```python
from keras.models import Sequential
from keras.layers import Dense
ann=Sequential()

ann.add(Dense(units=23,activation='relu'))
ann.add(Dense(units=100,activation='relu'))
ann.add(Dense(units=50,activation='relu'))
ann.add(Dense(4,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
np.unique(y_train)
```

```python
import tensorflow as tf
tf_callback=tf.keras.callbacks.TensorBoard(log_dir='logs/',histogram_freq=1)
```
