Face detection in a vedio
---

```python
# video - collection of images
import cv2
```

```python
video=cv2.VideoCapture(0)
while True:
    success,img=video.read()
    
    cv2.imshow('Frame',img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break

video.release() #video quit
cv2.destroyAllWindows()
print(success)
```
