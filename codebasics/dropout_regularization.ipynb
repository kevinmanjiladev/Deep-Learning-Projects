Imagine you are in a submarine. You send a sound wave (sonar), and it hits something underwater. The echo that comes back contains patterns.

# ğŸ” **Why 60 features?**

Sonar pulses return echoes over time. Engineers divided the returning signal into **60 time windows**, each measuring the **energy** in that window.

Think of it like:

```
Time window 1:   low energy
Time window 2:   medium energy
...
Time window 60:  high energy
```

This â€œshapeâ€ helps distinguish mines vs rocks.
